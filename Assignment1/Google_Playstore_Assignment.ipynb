```python
# Replacing
my_df["Reviews"] = my_df["Reviews"].str.replace('3.0M','3000000')
```

```python
def convert_size(size):
    if 'k' in size:
        return float(size[:-1])
    elif 'M' in size:
        return float(size[:-1]) * 1000
    elif 'Varies with device' in size:
        return np.nan
    else:
        return np.nan
```

```python
my_df['Size'] = my_df['Size'].apply(convert_size)
```

```python
my_df.head()
```

```python
my_df.info()
```

```python
my_df['Installs'].unique()
```

```python
cols = ['Installs', 'Price']
chars_to_remove = ['+', ',', '$']
translation_table = str.maketrans('', '', ''.join(chars_to_remove))
for col in cols:
    my_df[col] = my_df[col].str.translate(translation_table)
      
```

```python
my_df.info()
```

```python
my_df['Installs'] = my_df['Installs'].astype('int')
my_df['Price'] = my_df['Price'].astype('float')
```

```python
my_df.info()
```

```python
my_df['Last Updated'] = pd.to_datetime(my_df['Last Updated'])
my_df['day'] = my_df['Last Updated'].dt.day
my_df['month'] = my_df['Last Updated'].dt.month
my_df['year'] = my_df['Last Updated'].dt.year
```

Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
plt.figure(figsize=(16,10))
sns.boxplot(x='Category', y='Rating', data=my_df)
plt.xticks(rotation=90)
plt.title('Distribution of App Ratings by Category')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```

Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
average_price = my_df.groupby('Type')['Price'].mean()
plt.figure(figsize=(12,10))
average_price.plot(kind='bar', color='skyblue')
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
# plt.xticks(rotation=0)
plt.show()
```

Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
app_freq = my_df.groupby('App')['Installs'].sum().reset_index()
sorted_app_freq = app_freq.sort_values(by='Installs', ascending=False)
top_10_app = sorted_app_freq.head(10)
top_10_app
```

Q15. A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?

```python
"""
I would analyze Category, Installs, Reviews, Rating, Price, Size and Last updated features to identify the most popular app
categories. 
"""
```

```python
category_installs = my_df.groupby('Category')['Reviews'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
category_installs.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 App Categories by Reviews')
plt.xlabel('Category')
plt.ylabel('Reviews')
plt.xticks(rotation=75)
plt.show()
```

Q16. A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
"""
I would analyze Apps, Installs, Rating, Reviews and Category features. For Visualization I can use Bar chart and scatter plot.
"""
```

```python
developer_metrics = my_df.groupby('Android Ver').agg({'Rating': 'mean', 'Installs': 'sum'}).sort_values(by='Installs', ascending=False)
plt.figure(figsize=(10, 6))
developer_metrics.head(15).plot(kind='bar', y='Installs', color='salmon')
plt.title('Top 15 Android Version by Total Installs')
plt.xlabel('Version')
plt.ylabel('Total Installs')
plt.xticks(rotation=90)
plt.show()
```

Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
"""
I would analyze last updated, Installs, Rating and Reviews over the time features. For visualization I would use line chart or Bar charts.
"""
```

```python
my_df['Last Updated date'] = pd.to_datetime(my_df['Last Updated'])
my_df['Month_Year'] = my_df['Last Updated date'].dt.to_period('M')

monthly_metrics = my_df.groupby('Month_Year').agg({'Installs': 'sum', 'Rating': 'mean'})

fig, ax1 = plt.subplots(figsize=(12, 6))

color = 'skyblue'
ax1.set_xlabel('Month and Year')
ax1.set_ylabel('Total Installs', color=color)
ax1.bar(monthly_metrics.index.astype(str), monthly_metrics['Installs'], color=color)
ax1.tick_params(axis='x', labelrotation=90)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  
color = 'salmon'
ax2.set_ylabel('Average Rating', color=color)  
ax2.plot(monthly_metrics.index.astype(str), monthly_metrics['Rating'], color=color, marker='o')  # Using markers for better visibility
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Monthly Installs and Average Rating Trend')
fig.tight_layout()
plt.show()
```
